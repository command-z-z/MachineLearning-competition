```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
x, y = data.iloc[:,:-1], data.iloc[:,-1]
```

```python
matrix = xgb.DMatrix(data = x, label = y)
```

### 训练集/评测集的制作

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

### New一个Regressor

#### 超参

- learning_rate：学习率(0,1)
- max_depth：限制树的最大深度
- colsample：列采样率
- subsample：每棵树的样本采样率
- n_etstimators：子树的数量
- bjective：树的学习类型
 - 'reg:linear'：线性回归
 - 'reg:logistic'：二分类
 - 'gamma/alpha：L1/L2'

```python
model = xgb.XGBRegressor(objective = 'reg:linear', colsample_bytree = 0.7, learning_rate = 0.05, max_depth = 10, n_estimators = 800)
```

### fit(x_train,y_train)

```python
model.fit(x_train,y_train)
```

### 评测过程
1. predict(x_test) 得到评测结果
2. 与真正的测试集y_test进行比较，比较函数使用mse或者loggedmse（数字特别大就用loggedmse）

```python
preds = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, preds)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
loggedrmse = np.sqrt(mean_squared_error(np.log(y_test), np.log(preds)))
loggedrmse
```

### 调参
- n_estimator：适当增加，过大可能会过拟合
- colsample_bytree：适当减小
- learning_rate；适当减小
