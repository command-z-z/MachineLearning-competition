## 数据的简单填充

### Numerical
1. 均值填充
2. 中位数

### Categorial填充
1. 控制填充【增加一个"None"-"NA"】

2. 高频填充

3. 关系填充【根据数据的含义，参考其他的列进行填充】

```python
train1 = train1.fillna("None")
```

```python
def getNullCount(t):
    enpty = t.isnull().sum()
    print(enpty[enpty > 0])
```

## 训练集的制作

```python
y = train1['SalePrice']
train1 = train1.drop(['Id' ,'SalePrice'], axis = 1)
```

```python
train1.head(10)
```

```python
x = pd.get_dummies(train1).reset_index(drop = True) # one_hot
```

```python
x.head(10)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
matrix = xgb.DMatrix(data = x,label = y)
```

```python
# 数据集进行拆分，80%训练，20%用来测试
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 123)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# XGBoost
xg_reg = xgb.XGBRegressor(objective = 'reg:linear',colsample_bytree = 0.6,learning_rate = 0.01,max_depth = 8,alpha = 10, n_estimators = 700, subsample = 0.7)
```

```python
xg_reg.fit(x_train,y_train)
```

```python
pred = xg_reg.predict(x_test)
```

```python
pred
```

```python
rmse = np.sqrt(mean_squared_error(np.log(y_test),np.log(pred)))
```
