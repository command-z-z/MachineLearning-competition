## 结构化数据散点图

```python
data = {
    'a' : np.arange(50),
    'c' : np.random.randint(0,50,50),
    'd' : np.random.randn(50)
}
data
```

```python
plt.scatter(data['a'],data['d'])
plt.xlabel('a 数据')
plt.ylabel('b 数据')
```

```python
data['d'] = np.abs(data['d'])
```

## 散点图的参数
1. c：color 'r'/'red' 'g'/'green'
2. marker：-- -. > ^

```python
plt.scatter('a','d', data = data,c = 'c', marker = '>')
```

## 柱状图绘制

```python
names = ['a类型','b类型','c类型']
values = [1,10,100]
```

```python
plt.bar(range(len(names)),values)
plt.xticks(range(len(names)),names)
```

```python
plt.scatter(range(len(names)),values)
plt.xticks(range(len(names)),names)
plt.title('离散数据散点图')
```

## 子图绘制--SubPlot
1. 将画布切分
2. 将plot图分配到固定的位置和大小

```python
plt.figure(1, figsize=(15,5))

plt.subplot(131)# rcn row columns number
plt.bar(range(len(names)),values, color = 'r')
plt.xticks(range(len(names)),names)

plt.subplot(232)
plt.scatter(range(len(names)),values, color = 'y')
plt.xticks(range(len(names)),names)

plt.subplot(235)
plt.plot(range(len(names)),values)
plt.xticks(range(len(names)),names)
plt.suptitle('离散数据的绘图')
```

## Seaborn绘图

基于Pyplot，提供更多支持

自带iris,titanic...

图形绘制进行优化

提供多种可供参考的高级绘图--带状图，蜂群图，箱型图，提琴图，分布图

```python
import seaborn as sns
```

```python
# 加载数据集
tips = sns.load_dataset('tips')
tips
```

### 1.带状图 离散数据与连续数据之间的关系

```python
sns.stripplot(data = tips,x = 'day',y='total_bill')
```

```python
# 使用jitter让数据左右抖动
sns.stripplot(data = tips,x = 'day',y='total_bill',jitter=True) 
```

### 2.蜂群图

```python
sns.swarmplot(x = 'day',y = 'total_bill',data=tips)
```

#### Hue --分组参数

```python
sns.swarmplot(x = 'day',y = 'total_bill',data=tips,hue = 'time')
```

```python
sns.swarmplot('day','total_bill',data=tips,hue = 'sex')
```

```python
sns.swarmplot('day','total_bill',data=tips,hue = 'size')
```

```python
sns.swarmplot('size','total_bill',data = tips)
```

```python
tips['size'].corr(tips['total_bill'])
```

### 3.箱线图

```python
sns.boxplot('day','total_bill',data = tips)
```

```python
sns.boxplot('day','total_bill',data = tips,hue='time')
```

### 4.提琴图 violinplot()

```python
sns.violinplot('day','total_bill',data = tips,hue='time',split=True)
```

### 多图合成一个图展示

```python
sns.violinplot('day','total_bill',data = tips)
sns.swarmplot('day','total_bill',data = tips,color = 'w')
```
